```python
import pandas as pd

df1 = pd.read_excel('recipes_data.xlsx')
df2 = pd.read_excel('recipes_data2.xlsx')
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
df = df[df['Calories'] != 'Could not find calories']
df['Calories'] = df['Calories'].str.replace('Cals: ', '', regex=False)
df['Calories'] = df['Calories'].astype(float)

df = df[df['Points'] != 'Could not find points']
df['Points'] = df['Points'].astype(int)
df
```

# Step 3 – EDA and Visualisation

```python
def visualize(column, interval):
  # Calculate important values
  min_val = df[column].min()
  max_val = df[column].max()
  q1 = df[column].quantile(0.25)
  med = df[column].median()
  q3 = df[column].quantile(0.75)

  print(f"Minimum   {column}  : {min_val}")
  print(f"Maximum  {column} :  {max_val}")
  print(f"Q1 Calories: {q1}")
  print(f"Median Calories: {med}")
  print(f"Q3 Calories: {q3}")


  # Create a figure with three subplots
  fig, axes = plt.subplots(1, 3, figsize=(18, 5))

  # Histogram
  sns.histplot(df[column], bins=range(0, int(df[column].max()) + interval, interval), ax=axes[0])
  axes[0].set_title('Histogram of ' + column)
  axes[0].set_xlabel(column)
  axes[0].set_ylabel('Frequency')

  # Box plot
  sns.boxplot(y=df[column], ax=axes[1])
  axes[1].set_title('Box Plot of ' + column)
  axes[1].set_ylabel(column)

  # Violin plot
  sns.violinplot(y=df[column], ax=axes[2])
  axes[2].set_title('Violin Plot of ' + column)
  axes[2].set_ylabel(column)

  plt.tight_layout()
  plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

visualize("Calories",50)
```

```python
visualize("Points",1)
```

```python
from collections import defaultdict

# Create a dictionary to store the counts of each key
key_counts = defaultdict(int)
for keys_str in df['Keys']:
  if keys_str != 'Could not find keys':
    for key in keys_str.split(', '):
      key_counts[key] += 1

# Sort the keys by their counts in descending order
sorted_keys = sorted(key_counts.items(), key=lambda item: item[1], reverse=True)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(*zip(*sorted_keys))
plt.xlabel('Keys')
plt.ylabel('Frequency')
plt.title('Distribution of Keys')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Create a dictionary to store the counts of each key
key_counts = defaultdict(int)
for keys_str in df['Keys']:
  if keys_str != 'Could not find keys':
    for key in keys_str.split(', '):
      key_counts[key] += 1

# Convert the counts into a list of labels and values for the pie chart
labels = list(key_counts.keys())
sizes = list(key_counts.values())

# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Ratio of Keys in Recipes')
plt.show()
```

```python
# Calculate average calories for each key
key_avg_calories = {}
for index, row in df.iterrows():
  if row['Keys'] != 'Could not find keys':
    for key in row['Keys'].split(', '):
      if key not in key_avg_calories:
        key_avg_calories[key] = {'total_calories': 0, 'count': 0}
      key_avg_calories[key]['total_calories'] += row['Calories']
      key_avg_calories[key]['count'] += 1

for key in key_avg_calories:
  key_avg_calories[key] = key_avg_calories[key]['total_calories'] / key_avg_calories[key]['count']

# Sort keys by average calories in descending order
sorted_keys_calories = sorted(key_avg_calories.items(), key=lambda item: item[1], reverse=True)

# Calculate average points for each key
key_avg_points = {}
for index, row in df.iterrows():
  if row['Keys'] != 'Could not find keys':
    for key in row['Keys'].split(', '):
      if key not in key_avg_points:
        key_avg_points[key] = {'total_points': 0, 'count': 0}
      key_avg_points[key]['total_points'] += row['Points']
      key_avg_points[key]['count'] += 1

for key in key_avg_points:
  key_avg_points[key] = key_avg_points[key]['total_points'] / key_avg_points[key]['count']

# Sort keys by average points in descending order
sorted_keys_points = sorted(key_avg_points.items(), key=lambda item: item[1], reverse=True)

# Create subplots for visualization
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Average calories per key
axes[0].bar(*zip(*sorted_keys_calories))
axes[0].set_xlabel('Keys')
axes[0].set_ylabel('Average Calories')
axes[0].set_title('Average Calories per Key')
axes[0].tick_params(axis='x', rotation=90)

# Average points per key
axes[1].bar(*zip(*sorted_keys_points))
axes[1].set_xlabel('Keys')
axes[1].set_ylabel('Average Points')
axes[1].set_title('Average Points per Key')
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
avg_calories_by_point = df.groupby('Points')['Calories'].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_calories_by_point.index, avg_calories_by_point.values, marker='o')
plt.xlabel('Points')
plt.ylabel('Average Calories')
plt.title('Average Calories for Each Point')
plt.grid(True)
plt.show()
```

# Step 4 – User interaction

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Define widgets for calorie and points ranges
calorie_range_slider = widgets.IntRangeSlider(
    value=[df['Calories'].min(), df['Calories'].max()],
    min=df['Calories'].min(),
    max=df['Calories'].max(),
    step=1,
    description='Calorie Range:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
)

points_range_slider = widgets.IntRangeSlider(
    value=[df['Points'].min(), df['Points'].max()],
    min=df['Points'].min(),
    max=df['Points'].max(),
    step=1,
    description='Points Range:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
)

# Define a button to trigger the filtering
filter_button = widgets.Button(description="Filter Recipes")

# Create an output widget to display the filtered DataFrame
output = widgets.Output()

def filter_recipes(b):
    with output:
        clear_output()
        calorie_min = calorie_range_slider.value[0]
        calorie_max = calorie_range_slider.value[1]
        points_min = points_range_slider.value[0]
        points_max = points_range_slider.value[1]

        filtered_df = df[(df['Calories'] >= calorie_min) & (df['Calories'] <= calorie_max)]

        if 'Points' in df.columns:
            filtered_df = filtered_df[(filtered_df['Points'] >= points_min) & (filtered_df['Points'] <= points_max)]
            df_sorted = filtered_df.sort_values('Calories', ascending=False)
            top_10_highest_calories = df_sorted.head(10)
        # display(top_10_highest_calories)
        num = 1
        for index, row in top_10_highest_calories.iterrows():
          food_name = row['Food Name']
          calories = row['Calories']
          points = row['Points']
          image_link = row['Image Link']
          summary = row['Summary']

          html_code = f"""
          <div>
            <h3>{num}. {food_name}, {calories} calories, {points} WW points.</h3>
            <p>{summary}</p>
            <img src="{image_link}" width="100">

          </div>
          """
          num+=1
          display(HTML(html_code))


filter_button.on_click(filter_recipes)
widgets.VBox([calorie_range_slider, points_range_slider, filter_button, output])
```
